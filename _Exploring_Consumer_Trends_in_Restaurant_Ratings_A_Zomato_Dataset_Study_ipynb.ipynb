```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
resturants= pd.read_csv('zomato.csv',encoding = "ISO-8859-1")
countrycode=pd.read_excel('Country-Code.xlsx')
```

```python
resturants=resturants.merge(countrycode,on='Country Code')
```

```python
resturants.head()
```

```python
indiaRes = resturants[resturants['Country']=='India']
singaporeRes = resturants[resturants['Country']=='Singapore']
australiaRes = resturants[resturants['Country']=='Australia']
brazilRes = resturants[resturants['Country']=='Brazil']
canadaRes = resturants[resturants['Country']=='Canada']
UAERes = resturants[resturants['Country']=='UAE']
UKRes = resturants[resturants['Country']=='UK']
USRes = resturants[resturants['Country']=='US']
```
